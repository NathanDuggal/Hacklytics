```python
import json
import dateparser
import numpy as np
trends, songs, weather = 0,0,0
with open("weathery.csv", "r") as weather, open("yeast.json", 'r') as trends, open("tracks_features.csv",'r') as songs:
    trends = json.loads(trends.read())
    songs = songs.read()
    weather = weather.read()
```

```python

import pandas as pd

weather_dict = {}
X,Y = 0,0

for line in weather.split('\n')[1:]:
    line = line.split(',')
    line = [e if e!='' else '0' for e in line]
    weather_dict[line[1]] = list(map(float,line[2:]))

songs_dict = {}

for line in songs.split('\n')[1:]:
    line_end = line.split("True,")[-1] if 'True' in line else line.split("False,")[-1]
    line = line.split(",")
    
    songs_dict[line[1]] = line_end.split(',')[:-1]

    #print(songs_dict[line[1]])

inputs = []
predicting = []

for song in trends:
    dates = trends[song]
    arr = [dates[e][0] for e in dates.keys()]
    if arr!=[]:
        cutoff = np.percentile(arr, 75)
        rows = [[weather_dict[dateparser.parse(e).strftime("%Y-%m-%d")],[float(j) for j in songs_dict[song]]] for e in dates if dates[e][0]>cutoff]
        inputs += [e[0] for e in rows]
        predicting += [e[1] for e in rows]
        
X = pd.DataFrame(data=inputs)
Y = pd.DataFrame(data=predicting)

X.
    
```
