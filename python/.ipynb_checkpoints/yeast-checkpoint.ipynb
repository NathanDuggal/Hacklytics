```python
import json
import numpy as np
trends, songs, weather = 0,0,0
with open("weathery.csv", "r") as weather, open("yeast.json", 'r') as trends, open("track_feature_output.csv",'r') as songs:
    trends = json.loads(trends.read())
    songs = songs.read()
    weather = weather.read()
```

```python
new_dict = {}

songs_dict = {}
for line in songs.split('\n'):
    line = line.split(",")
    songs_dict[line[0]] = line[2:]

for song in trends:
    dates = trends[song]
    arr = [dates[e][0] for e in dates.keys()]
    cutoff = np.percentile(arr, 75)
    print([[e,songs_dict[song]] for e in dates if dates[e][0]>cutoff])
    
    
```
